## Mutable vs Immutable

### Z≈Ça wersja

```python
def printing(lista = []):
    lista.append('print')
    for el in lista:
        print(id(lista), el)
        
printing()
```

```python
l = ['nowe s≈Çowo']
print(id(l))
printing(l)
```

```python
printing()
```

### Dobra wersja

```python
def printing(lista = None):
    print("Przed if: ",id(lista))
    lista = lista if lista else [] # druga opcja lista or []
    print("Po if: ", id(lista))
    lista.append('print')
    for el in lista:
        print(id(lista), el)
```

```python
printing()
```
